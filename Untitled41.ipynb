```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
plt.style.use('fivethirtyeight')

x=np.linspace(-10,10,100)

def plot_graph(x,y,alpha=0.6,
               label_x = r"$x \rightarrow$", label_y=r"$act(x) \rightarrow$",
               title=None,
               LABEL=None):
    plt.figure(figsize=(10,7))
    plt.axhline(y=0,color='black',linestyle='--',lw=2)#X axis
    plt.axvline(x=0,color='black',linestyle='--',lw=2)#Y axis
    plt.xlabel(label_x)
    plt.ylabel(label_y)

    if LABEL !=None:
        plt.plot(x,y,alpha=alpha,label=LABEL)
        plt.legend()
      
    else:
        plt.plot(x,y,alpha=alpha)

def derative(f,x,delta_x=1e-6):
  return (f(x+delta_x)-f(x))/(delta_x)
```

```python
def sigmoid(x):
  return tf.keras.activations.sigmoid(x)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
plot_graph(x,sigmoid(x),title='Sigmoid function',
           LABEL=r"$\sigma(x) = \frac{1}{1+e^{x}}$")
plot_graph(x,derative(sigmoid,x),title='Derivative of SigmoidFunction',
           LABEL=r"$\sigma^\prime(x) = \sigma(x)(1-\sigma(x))$")
```

```python
def relu(x):
    return tf.keras.activations.relu(x, alpha=0.0, max_value=None, threshold=0)
```

```python
def relu(x):
    return np.where(x>=0, x, 0)
```

```python
plot_graph(x, relu(x), title="ReLU",
           LABEL=r"$ReLU(x)= max(x,0)$");

plot_graph(x, derative(relu, x), title="Derivative of ReLU function", 
           LABEL=r"$ReLU^\prime(x)$")
```

```python
def leaky_relu(x, alpha=0.3):
    return np.where(x>=0, x, alpha*x)
```

```python
plot_graph(x, leaky_relu(x), title="Leaky ReLU");

plot_graph(x, derative(leaky_relu, x), title="Derivative of LeakyReLU function", 
           LABEL=r"$Leaky\_ReLU^\prime(x)$")
```
